# AB Test Analysis Mobile Payment Screen Redesign

## Overview
This project analyzes an AB test conducted on a mobile payment screen redesign.
The goal was to determine whether the new design increases the number of successful payments.

## Business Question
> Should the new payment screen design be implemented for all users, or rejected?

## Experiment Setup
| | Details |
|---|---|
| **Start date** | July 24, 2021 (00:00) |
| **Platform** | Mobile only |
| **Control (Group 0)** | Old payment screen |
| **Test (Group 1)** | New payment screen |

## Methodology
1. **Filter** — Mobile users registered on or after July 24 (experiment start)
2. **User-level aggregation** — A user is "converted" if they made at least 1 successful payment
3. **Conversion Rate** — Unique converted users / Total unique users per group
4. **Statistical tests** — Chi-square, Z-test, 95% Confidence Interval

## Key Decisions & Assumptions
| Question | Decision | Reason |
|---|---|---|
| Filter by `date_reg` or `date_payment`? | `date_reg` | Users assigned to experiment at registration |
| Include "other" platform? | No | Experiment scoped to mobile only |
| CR denominator? | Total unique users in split | Goal is to increase payments across all users |
| Multiple payments per user? | Take `max()` | User counts as converted if succeeded at least once |

## Results
| Metric | Control | Test |
|---|---|---|
| Users | 5,877 | 5,915 |
| Conversions | 171 | 171 |
| Conversion Rate | 2.91% | 2.89% |
| Difference | — | -0.02 pp |

## Statistical Evidence
| Test | Result | Conclusion |
|---|---|---|
| Chi-square | p = 0.9955 | Not significant |
| Z-test | p = 0.9484 | Not significant |
| 95% CI | [-0.626%, +0.586%] | Includes zero |

## Final Decision
**Reject the new design.**

The new payment screen does not statistically improve conversion rates.
The observed difference is consistent with random variation.
The innovation should **not** be rolled out to all users.
